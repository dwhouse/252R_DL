```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader

documents = SimpleDirectoryReader("data").load_data() # data 폴더 내의 모든 문서 로드
index = VectorStoreIndex.from_documents(documents)
```

```python
query_engine = index.as_query_engine() # 쿼리 엔진 설정
response = query_engine.query("작가의 청소년 시절은 어떠했나요?") # 요청
print(response) # 응답 출력
```
